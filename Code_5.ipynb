```python
import pandas as pd

df = pd.read_stata(r"D:\NFHS\DataBase-Files\2019-2021\Individual Recode-IAIR7EDT\IAIR7EFL.DTA", convert_categoricals=False)
```

```python
thresholds = {
    'v445': {'label': 'BMI', 'unit': '', 'cutoff': 18.5, 'comparison': 'less_than'},
    'sb29s': {'label': 'Systolic_BP', 'unit': 'mmHg', 'cutoff': 130, 'comparison': 'greater_than'},
    'sb18s': {'label': 'Glucose', 'unit': 'mg/dL', 'cutoff': 140, 'comparison': 'greater_than'},
    'sb74': {'label': 'Waist', 'unit': 'cm', 'cutoff': 80, 'comparison': 'greater_than'},
    'sb15': {'label': 'Hemoglobin', 'unit': 'g/dL', 'cutoff': 12.0, 'comparison': 'less_than'}
}
#sources for these thresholds are all from valid papers and WHO and ACC/AHA datas ( doub check will be neccessary ) 
```

```python
# show abronamities based on these threhsolds
for var, info in thresholds.items():
    label = info['label']
    cutoff = info['cutoff']
    comp = info['comparison']
    
    if comp == 'greater_than':
        df[f'{label}_abnormal'] = df[var] >= cutoff
    elif comp == 'less_than':
        df[f'{label}_abnormal'] = df[var] < cutoff
```

```python
df[['BMI_abnormal', 'Systolic_BP_abnormal', 'Glucose_abnormal', 'Waist_abnormal', 'Hemoglobin_abnormal']].head()
```

```python
# totally normal (all False = normal)
df['All_Normal'] = ~(df['BMI_abnormal'] | df['Systolic_BP_abnormal'] | df['Glucose_abnormal'] | df['Waist_abnormal'] | df['Hemoglobin_abnormal'])

# number of respondents
total = len(df)

# Number and percent of normal women
n_normal = df['All_Normal'].sum()
percent_normal = (n_normal / total) * 100


print(f"Totally normal women: {n_normal} out of {total} ({percent_normal:.2f}%)")
```
