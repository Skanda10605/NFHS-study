```python
def binarize(var,classes,text,table = {}):
    if text != 'time since last period':
        for i in range(len(classes) - 1):
            lb = classes[i][0]
            ub = classes[i][1]
            name = f'{text} {lb} to {ub}'
            table[name] = 0
            table[name] += ((pdf[var] >= lb) &
                                      (pdf[var] < ub)).astype(int)
        lb = classes[-1][0]
        name = f'{text} >= {lb}'
        table[name] = 0
        table[name] += (pdf[var] >= lb).astype(int)
        
        return table
    else:
        for i in range(len(classes) - 1):
            lb = classes[i][0]
            ub = classes[i][1]
            name = f'{text} {lb%100} months to {ub%100} months'
            table[name] = 0
            table[name] += ((pdf[var] >= lb) &
                                      (pdf[var] < ub)).astype(int)
        lb = classes[-1][0]
        name = f'{text} >= {lb%100} months'
        table[name] = 0
        table[name] += (pdf[var] >= lb).astype(int)
        return table
        
```

```python
def short_interval(pdf):
    # Define age bins
    age_bins = ["<20", "20 - 30", "30 - 40", "40 - 50"]
    result = pd.DataFrame(index=pdf.index, columns=[f"short interval at age {bin}" for bin in age_bins], dtype=int)

    # Extract relevant columns
    children = pdf["v201"]  # Number of children
    start_age = pdf["v212"]  # Mother's age at first birth

    # Create interval matrix (assuming max 20 children to avoid iterating)
    interval_cols = [f"b11_{k:02d}" for k in range(1, 21) if f"b11_{k:02d}" in pdf.columns]
    intervals = pdf[interval_cols].fillna(0).to_numpy()

    # Vectorized processing
    valid_intervals = (intervals > 0) & (intervals < 33)  # Short intervals < 33 months
    ages = np.repeat(start_age.values[:, np.newaxis], intervals.shape[1], axis=1)
    ages += np.cumsum(intervals / 12, axis=1)  # Compute ages at each birth

    # Assign to bins
    result.iloc[:, 0] = np.sum(valid_intervals & (ages < 20), axis=1)
    result.iloc[:, 1] = np.sum(valid_intervals & ((ages >= 20) & (ages < 30)), axis=1)
    result.iloc[:, 2] = np.sum(valid_intervals & ((ages >= 30) & (ages < 40)), axis=1)
    result.iloc[:, 3] = np.sum(valid_intervals & ((ages >= 40) & (ages < 50)), axis=1)

    return result
```
